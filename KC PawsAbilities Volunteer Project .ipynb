### Number of Likes Over Time

```python
#number of likes based on month 

plt.figure(figsize=(8,6))
df['likes '].value_counts(normalize=True).plot.bar()
plt.title('Distribution of Title Ratings')
plt.xlabel('month')
plt.ylabel('relative frequency')

plt.show()
```

```python
#number of post per month 

df.groupby('month').size()
```

***The highest number of likes are typically received in the months of March, June, February, and December. You may notice that there tends to be more posted during these months which would explain the higher number of likes.***
